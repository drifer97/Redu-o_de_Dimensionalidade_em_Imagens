```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Carregue a imagem colorida
imagem_colorida = cv2.imread('/content/Nona.png')
```

```python
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)
```

```python
# Binarize a imagem (threshold)
limiar = 127  # O limiar para binarização
_, imagem_binarizada = cv2.threshold(imagem_cinza, limiar, 255, cv2.THRESH_BINARY)
```

```python
cv2_imshow(imagem_colorida)
cv2_imshow(imagem_cinza)
cv2_imshow(imagem_binarizada)
```

```python
# Salve a imagem binarizada em um arquivo
cv2.imwrite('imagem_binarizada.jpg', imagem_binarizada)
```

```python
# Aguarde uma tecla e depois feche as janelas
cv2.waitKey(0)
cv2.destroyAllWindows()
```
